```python
import pandas as pd
import seaborn as sns
```

```python
X_train = pd.read_csv('xtrain.csv')
X_test = pd.read_csv('xtest.csv')

y_train = pd.read_csv('ytrain.csv')
y_test = pd.read_csv('ytest.csv')
```

```python
X_train.head()
```

# The Target Variable

```python
y_train.value_counts()
```

```python
def convert_y(value):
    target_dict = {'Mitochondrial genetic inheritance disorders': 0, 'Single-gene inheritance diseases': 1, 'Multifactorial genetic inheritance disorders': 2}
    
    return target_dict[value]
```

```python
y_train = y_train['Genetic Disorder'].map(convert_y)
y_test = y_test['Genetic Disorder'].map(convert_y)
```

```python
# The Target Value
from tensorflow.keras.utils import to_categorical

y_cat_train = to_categorical(y_train.values, 3)
y_cat_test = to_categorical(y_test.values, 3)
```

# Model Training

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
```

```python
model  = Sequential()

model.add(Dense(43, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(22, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(11, activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(3, activation='softmax'))

model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

model.fit(x = X_train.values, y = y_cat_train, epochs = 50, batch_size = 256, validation_data=(X_test.values, y_cat_test))
```

```python
losses = pd.DataFrame(model.history.history)
```
