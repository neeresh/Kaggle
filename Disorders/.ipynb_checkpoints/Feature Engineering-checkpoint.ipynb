```python
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
df1 = pd.read_csv('train_genetic_disorders.csv')
print(df1.shape)

df1.head()
```

```python
df1 = df1.drop(['Patient Id', 'Patient First Name', 'Family Name', "Father's name"], axis = 1)
```

```python
print(df1.shape)
df1.head()
```

# The Target Variable

```python
df1['Genetic Disorder'].value_counts()
```

```python
df1 = df1.dropna(subset = ['Genetic Disorder'], axis = 0)
df1 = df1.reset_index()
df1.shape
```

```python
df2 = pd.read_csv('feature_engineering.csv')

print(df2.shape)
df2.head()
```

```python
df2.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df = pd.concat([df1, df2], axis = 1)
```

```python
print(df.shape)

df.tail()
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(['Genetic Disorder'], axis = 1),
                                                   df['Genetic Disorder'],
                                                   test_size=0.3,
                                                   random_state=12)
```

# The Variables Type

```python
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O']
print("Total number of numerical features: ", len(numerical_features))

print(numerical_features)
```

```python
categorical_features = [feature for feature in df.columns if feature not in numerical_features]
print("Total number of categorical features: ", len(categorical_features))

print(categorical_features)
```

# Numerical Features

## Feature Engineering

```python
X_train['no_of_tests'] = X_train[['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5']].sum(axis = 1)
X_test['no_of_tests'] = X_test[['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5']].sum(axis = 1)

X_train = X_train.drop(['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5'], axis = 1)
X_test = X_test.drop(['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5'], axis = 1)
```

```python
X_train['no_of_symptoms'] = X_train[['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5']].sum(axis = 1)
X_test['no_of_symptoms'] = X_test[['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5']].sum(axis = 1)

X_train = X_train.drop(['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5'], axis = 1)
X_test = X_test.drop(['Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5'], axis = 1)
```

```python
X_train.head()
```

```python
X_test.head()
```

## Missing Features

```python
numerical_features = [feature for feature in X_train.columns if X_train[feature].dtype != 'O']
print("Total number of numerical features: ", len(numerical_features))

print(numerical_features)
```

```python
missing_features = [feature for feature in X_train.columns if X_train[feature].isnull().sum() > 0]
print("Total number of missing features: ", len(missing_features))

print(missing_features)
```

```python
numerical_features_na = [feature for feature in missing_features if X_train[feature].dtype != 'O']
print("Total number of numerical missing features: ", len(numerical_features_na))

print(numerical_features_na)
```

```python
categorical_features_na = [feature for feature in missing_features if feature not in numerical_features_na]
print("Total number of categorical missing features: ", len(categorical_features_na))

print(categorical_features_na)
```

### Numerical Features

```python
X_train[numerical_features_na].isnull().mean().sort_values(ascending = False)
```

#### Random Sample Imputation

```python
from feature_engine.imputation import RandomSampleImputer
```

```python
random_sample_imputer = ['Patient Age', "Mother's age", "Father's age", 'No. of previous abortion', 'White Blood cell count (thousand per microliter)']

random_sample = RandomSampleImputer(variables=random_sample_imputer, random_state=22)
random_sample.fit(X_train)

X_train = random_sample.transform(X_train)
X_test = random_sample.transform(X_test)
```

```python
X_train[numerical_features].isnull().sum()
```

#### Discretisation

```python
ewd_b3 = ['Patient Age']
ewd_b4 = ['No. of previous abortion']
efd_q3 = ["Mother's age", "Father's age"]
efd_q14 = ['White Blood cell count (thousand per microliter)']
```

```python
from feature_engine.discretisation import EqualWidthDiscretiser

disc_b4 = EqualWidthDiscretiser(bins=4, variables = ewd_b4)
disc_b4.fit(X_train)

X_train = disc_b4.transform(X_train)
X_test = disc_b4.transform(X_test)
```

```python
disc_b3 = EqualWidthDiscretiser(bins=3, variables = ewd_b3)
disc_b3.fit(X_train)

X_train = disc_b3.transform(X_train)
X_test = disc_b3.transform(X_test)
```

```python
from feature_engine.discretisation import EqualFrequencyDiscretiser

disc_q3 = EqualFrequencyDiscretiser(q=3, variables = efd_q3)
disc_q3.fit(X_train)

X_train = disc_q3.transform(X_train)
X_test = disc_q3.transform(X_test)
```

```python
disc_q14 = EqualFrequencyDiscretiser(q = 14, variables = efd_q14)
disc_q14.fit(X_train)

X_train = disc_q14.transform(X_train)
X_test = disc_q14.transform(X_test)
```

## Institute Name

```python
from feature_engine.imputation import AddMissingIndicator

missing_indicator = AddMissingIndicator(variables = ['Institute Name'])
missing_indicator.fit(X_train)

X_train = missing_indicator.transform(X_train)
X_test = missing_indicator.transform(X_test)
```

```python
X_train.head()
```

```python
X_train['Institution_Applicability_na'] = np.where(X_train['Institute Name'] == 'Not applicable', 'Not applicable', np.where(X_train['Institute Name'].isna(), np.nan, 'Applicable'))
X_test['Institution_Applicability_na'] = np.where(X_test['Institute Name'] == 'Not applicable', 'Not applicable', np.where(X_test['Institute Name'].isna(), np.nan, 'Applicable'))

X_train = X_train.drop(['Institute Name'], axis = 1)
X_test = X_test.drop(['Institute Name'], axis = 1)

X_train['Institution_Applicability_na'] = np.where(X_train['Institution_Applicability_na'] == 'nan', 'Missing', X_train['Institution_Applicability_na'])
X_test['Institution_Applicability_na'] = np.where(X_test['Institution_Applicability_na'] == 'nan', 'Missing', X_test['Institution_Applicability_na'])
```
