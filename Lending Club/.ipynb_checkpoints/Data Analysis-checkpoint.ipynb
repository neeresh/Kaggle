```python
df.shape
```

```python
def plot_hist(df, feature):
    fig, ax = plt.subplots(1, 2, figsize=(15, 5))

    sns.histplot(df[feature], ax = ax[0], kde = False, bins=30)
    ax[0].set_title('Bins')

    sns.histplot(df[feature], kde = True, ax = ax[1], bins = 30)
    ax[1].set_title('Kde')
```

```python
for feature in numerical_features:
    plot_hist(df, feature)
```

```python
print(numerical_features)
```

```python
skewed_features = ['loan_amnt', 'int_rate', 'installment', 'open_acc', 'revol_bal', 'total_acc', 'mort_acc']
print("Total number of skewed features: {}".format(len(skewed_features)))

print(skewed_features)
```
