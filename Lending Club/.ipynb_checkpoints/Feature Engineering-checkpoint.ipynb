### mort_acc

```python
# mort_acc column has 10% of missing data, and it represents the number of mort_acc each person has.
# Instead of removing missing rows from the data, we can remove this feature.

# or, we can replace the values in mort_acc using the correlated input feature.

X_train.corr()['mort_acc']
```

**Looks like mort_acc and total_acc are correlated better rest of the other.**

```python
X_train.groupby('total_acc').mean()['mort_acc']
```

```python
total_acc_avg_train = X_train.groupby('total_acc').mean()['mort_acc']
```

```python
def fill_mort_acc(total_acc, mort_acc):
    
    if np.isnan(mort_acc):
        return total_acc_avg_train[total_acc]
    else:
        return mort_acc
```

```python
X_train['mort_acc'] = X_train.apply(lambda x: fill_mort_acc(x['total_acc'], x['mort_acc']), axis = 1)
X_test['mort_acc'] = X_test.apply(lambda x: fill_mort_acc(x['total_acc'], x['mort_acc']), axis = 1)
```

```python
print(X_train['mort_acc'].isnull().sum())
print(X_test['mort_acc'].isnull().sum())
```
